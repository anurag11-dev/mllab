1. Create a dataframe for the following:

Area(in aana)   : [1 , 2, 2.5, 3, 4, 5 ]
Price(in Lakhs) : 1.2, 1.9, 2.3, 3.75, 4.8, 4.9

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Area(in aana)' : [1,2,2.5,3,4,5],
    'Price' : [1.2,1.9,2.3,3.75,4.8,4.9]
}
df = pd.DataFrame(data)
display(df.head())
df.describe()
```

a) Visualize the above dataset using scatterplot and find the pattern.

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Area(in aana)'],df['Price'])
plt.title('Area vs Price Scatter Plot')
plt.xlabel('Area(in aana)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

b) Apply Linear Regression and Find its equation

```python
from sklearn.linear_model import LinearRegression

X = df[['Area(in aana)']]
Y = df['Price']

model = LinearRegression()
model.fit(X,Y)
```

```python
#y=a+bx

a=model.intercept_
b=model.coef_
```

```python
print(f"Price = {a:.2f} + {b[0]:.2f}*Area")
```

c) Find MSE and R2-score for the given model

```python
y_pred = model.predict(X)
df['Predicted_Price']=y_pred
df.head()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(Y,y_pred)
r2 = r2_score(Y,y_pred)
print(f"MSE:{mse:.2f}")
print(f"R2-score:{r2:.2f}")
```

d) Plot the regression line in accordance with actual data

```python
X = df[['Area(in aana)']]
Y = df['Price']
plt.scatter(X,Y)
plt.plot(X,y_pred)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Land Price Visualization")
plt.grid(True)
plt.show()
```

e) Predict the Price for new data asked by the user

```python
Area = float(input('Enter the area value(in aana):'))
Y_new = model.predict([[Area]])
print(f"Predicted Price for {Area} aana is RS.{Y_new[0]:.2f}")
```

2. Create dataframe for the following data:

Years of Experience: 3 , 2, 4, 5, 3
Education: High School , Bachelor, High School, Bachelor, Master
Salary: 45000, 33000, 49025, 75000, 74000

```python
import pandas as pd
import numpy as np
data = {
    'Years_of_exp': [3, 2, 4, 5, 3],
    'Education': ['High School', 'Bachelor', 'High School', 'Bachelor', 'Master'],
    'Salary' : [45000, 33000, 49025, 75000, 74000]
}
df = pd.DataFrame(data)
display(df)
```

a) Encode the Education data first using either Label Encoding or One Hot Encoding Method

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
df['Education']=l.fit_transform(df['Education'])
df.head()
```

```python
X = df[['Years_of_exp','Education']]
Y = df['Salary']
```

b) Find the equation for this multiple linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)

a = model.intercept_
b = model.coef_

print(f'Salary = {a:.2f} + {b[0]:.2f} * Years_of_exp + {b[1]:.2f} * Education')
```

c) Find MSE and R2-score

```python
y_pred = model.predict(X)

df['Predicted_Price']=y_pred
df.head()

mse = mean_squared_error(Y, y_pred)

r2 = r2_score(Y, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")
```

d) Plot the reference line in accordance with the predicted data

```python
plt.scatter(Y,y_pred)
plt.plot([Y.min(),Y.max()],[Y.min(),Y.max()])
plt.show()
```
