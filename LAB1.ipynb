5. Calculate mean, median, mode, standard deviation and variance for a 1-D data. Also find the MSE and RMSE

```python
from scipy import stats
ExArray = np.array([1,2,3,4,5,5,5,5,3,4,4,3,2,1,1,2])
ExArray_pred = np.array([1,3,6,4,5,7,5,5,1,4,4,3,2,1,1,2])

print("Mean:",np.mean(ExArray))
print("Median:",np.median(ExArray))
print("Mode:",stats.mode(ExArray)[0])
print("Standard Deviation:",np.std(ExArray))
print("Variance:",np.var(ExArray))
errors = ExArray - ExArray_pred
print(errors)
mse = np.mean(errors ** 2)
print("MSE:",mse)
rmse = np.sqrt(mse)
print("RMSE:",rmse)
```

9. Create a line plot, scatter plot, bar graph and pie chart with labels using matplotlib for a given series of data

```python
import numpy as np
import matplotlib.pyplot as plt

subjects = ["Prob & Stats", "Data Sci & ML", "AI", "Comp Graphics", "Eng Economics", "Research Fund"]
marks = [60, 45, 59, 72, 80, 91]
x = np.arange(len(subjects))

# Line Plot
plt.plot(x, marks, marker="o")
plt.xticks(x, subjects, rotation=45)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Line Plot")
plt.show()

# Scatter Plot
plt.scatter(x, marks, color="tomato", s=100)
plt.xticks(x, subjects, rotation=45)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Scatter Plot")
plt.show()

# Bar Graph
plt.bar(subjects, marks, color="mediumseagreen")
plt.xticks(rotation=45)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Bar Graph")
plt.show()

# Pie Chart
plt.pie(marks, labels=subjects, autopct="%1.1f%%", startangle=140)
plt.title("Pie Chart")
plt.show()
```
